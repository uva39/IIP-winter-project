```python
import pandas as pd
import numpy as np
import random
import re
```

```python
df = pd.read_csv('raw_tags.csv')
df = df.T
df.columns = df.iloc[0]
df.drop(index='Type and category', inplace=True)
for col in df:
    df[col] = df[col].replace('\n','').str.strip()
df
```

```python
# Define the pattern to match
pattern = re.compile("^[a-zA-Z0-9_]+$")

def match_pattern(string):
    return np.nan if not pattern.search(string) else string

df = df.applymap(lambda x: match_pattern(x) if isinstance(x, str) else x)
df
```

```python
df = df.apply(lambda x: x.sort_values(key = lambda x: x.isna()).tolist())
df.dropna(inplace=True, thresh=1)
df = df.drop(df.columns[-5:], axis=1)
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv(r'../TagList.csv', index=False)
```
